```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=1.5, rc={'text.usetex': True})
sns.set_style("whitegrid")

from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
def plot_sd_theta(df):
    ax = sns.lineplot(data=pd.melt(df[['episode', r'$\theta^1$', r'$\theta^2$']], ['episode'], var_name='Probability of cooperation'),
              x='episode',
              y='value',
              hue='Probability of cooperation',
              style='Probability of cooperation'
              )

    ax.set(xlabel='Episode', ylabel=r'$\theta$')
    plt.savefig('NEW_2-player_simulation_theta.png', dpi=400)
    plt.show()
    plt.close()

def plot_sd_shares(df):
    ax = sns.lineplot(data=pd.melt(df[['episode', r'$m$', r'$n$']], ['episode'], var_name='Participation'),
              x='episode',
              y='value',
              hue='Participation',
              style='Participation'
              )

    ax.set(xlabel='Episode', ylabel='Shares owned')
    plt.savefig('NEW_2-player_simulation_shares.png', dpi=400)
    plt.show()
    plt.close()

def plot_sd_prices(df):
    ax = sns.lineplot(data=pd.melt(df[['episode', r'Price $m$', r'Price $n$']], ['episode'], var_name='Share prices'),
              x='episode',
              y='value',
              hue='Share prices',
              style='Share prices'
              )

    ax.set(xlabel='Episode', ylabel='Share prices')
    plt.savefig('NEW_2-player_simulation_prices.png', dpi=400)
    plt.show()
    plt.close()

def plot_sd_rewards(df):
    ax = sns.lineplot(data=df,
              x='episode',
              y='acc_rewards',
              )

    ax.set(xlabel='Episode', ylabel='Accumulated rewards')
    plt.savefig('NEW_2-player_simulation_rewards.png', dpi=400)
    plt.show()
    plt.close()

def get_env_data_test_multiple(csv):
    data = pd.read_csv(csv)
    df = pd.DataFrame(data)
    return df
```

```python
csv = 'simulation/2-player-simulation.csv'
df = get_env_data_test_multiple(csv)
df
```

```python
plot_sd_shares(df)
plot_sd_theta(df)
plot_sd_prices(df)

i = df[(df.episode == 0)].index
df = df.drop(i)
plot_sd_rewards(df)
```
